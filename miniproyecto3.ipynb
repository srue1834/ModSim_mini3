```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
n, p = 10, 0.5
x = np.arange(stats.binom.ppf(0, n, p), stats.binom.ppf(1, n, p))

plt.plot(x, stats.binom.pmf(x, n, p), label='binom pmf')
plt.vlines(x, 0, stats.binom.pmf(x, n, p), colors='g', lw=20, alpha=0.5)
plt.title('PMF')
plt.ylabel('probabilidad')
plt.xlabel('valores de variable aleatoria')
plt.show()
```
